### Task 3

Use order items data set and get total number of items sold as well as total revenue generated for a given product_id.
* Define a function which takes order_items and a product_id as arguments.
* product_id will be passed as integer
* Get number of items sold for a given product id using quantity (4th field)
* Compute revenue generated for a given product id using subtotal (5th field)
* Return the number of items sold as well as revenue generated. When we return more than one value, then the type of the returned results will be tuple.

````python
def get_product_metrics(order_items, product_id):
    product_metrics = (0,0.0)
    
    for order_item in orders_items :
        product_id = int(order_item.split(',')[2])
        order_metric = (int(order_item.split(',')[3]), float(order_item.split(',')[4]))
        if product_id == product_id:
              product_metrics = (product_metrics[0] + order_metric[0], product_metrics[1] + order_metric[1])
    return product_metrics
            
````

````python
def get_commission_amount(sales):
    commission_amount = 0.0
    for sale in sales:
        sale_amount = sale[0]
        commission_pct = round(sale[1]/100, 2) if len(sale) == 2 else 0
        commission_amount += sale_amount * commission_pct
    return round(commission_amount, 2)
````

````python
def get_count_by_order_status(orders):

    order_count = {}
    for order_status in orders:
        order_status_ = order_status.split(',')[3]
        if order_status_   in order_count :
            order_count[order_status_] +=1
        else :
            order_count[order_status_] =1
    return order_count
````

````python
##get revenue details per order 

def get_revenue_per_order(orders_items):
    revenue_per_order = {}

    for order_item in orders_items :
        order_item_order_id_ = int(order_item.split(',')[1])
        order_item_quanity = int(order_item.split(',')[3])
        order_item_subtotal = float(order_item.split(',')[4])
        order_metrics = {'order_item_quanity':order_item_quanity,'order_item_subtotal':order_item_subtotal}

        if revenue_per_order.get(order_item_order_id_):
            revenue_per_order[order_item_order_id_]['order_item_quanity'] += order_metrics['order_item_quanity']
            revenue_per_order[order_item_order_id_]['order_item_subtotal'] += order_metrics['order_item_subtotal']
        else:
            revenue_per_order[order_item_order_id_] =order_metrics
    return revenue_per_order
    ##this function should return a dict which contain order_item_order_id coressponding qunatity and order_revenu
    
````
